```python
 # Dependencies and Setup
import pandas as pd
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "momPOGstuff.csv"

# Read Purchasing File and store into Pandas data frame
pog_data = pd.read_csv(file_to_load)
```

```python
pog_data.head()
```

```python
import numpy as np
from itertools import chain

# return list from series of comma-separated strings
def chainer(s):
    return list(chain.from_iterable(s.str.split(',')))

# calculate lengths of splits
lens = pog_data['Store_num'].str.split(',').map(len)

# create new dataframe, repeating or chaining as appropriate
res = pd.DataFrame({'POD_ID': np.repeat(pog_data['POD_ID'], lens),
                    'Sign_Name': np.repeat(pog_data['Sign_Name'], lens),
                    'Store_num': chainer(pog_data['Store_num'])})

res


```

```python
res.to_csv('momsFinalPogSignStore.csv')
```
